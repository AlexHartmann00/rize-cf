```python
import json
import os
from typing import Iterable, Dict, Any

from google.cloud import firestore
from google.oauth2 import service_account

# TODO: optionally set project explicitly (otherwise taken from credentials)
PROJECT_ID = None  # e.g. "my-firebase-project-id"

# Path to your JSON file
JSON_PATH = "workouts.json"

# How many documents to write per batch (Firestore limit is 500)
BATCH_SIZE = 400


def load_json(path: str) -> Any:
    with open(path, "r", encoding="utf-8") as f:
        return json.load(f)

def main():
    # Firestore client
    if PROJECT_ID:
        db = firestore.Client(project=PROJECT_ID)
    else:
        creds = service_account.Credentials.from_service_account_file('rize-11838-firebase-adminsdk-fbsvc-ee0a388f16.json')
        db = firestore.Client(
            credentials=creds,
            project=creds.project_id
        )

    data = load_json(JSON_PATH)
    records = data['exercises']

    print(f"Loaded {len(records)} records from {JSON_PATH}")

    batch = db.batch()
    batch_count = 0
    total_written = 0

    for i, record in enumerate(records, start=1):
        doc_ref = (
            db.collection("workouts")
            .document()  # auto-ID
        )

        doc_ref.create(
            record
        )
        


if __name__ == "__main__":
    main()
```
